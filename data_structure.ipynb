### 1. Zbiór: inventory (Zasoby magazynowe)
Opis: Zawiera informacje o wszystkich zapasach w systemie, w tym wyrobach gotowych (FG) i półproduktach (SFG) oraz surowce (RM).
Kolumny:
- Item ID: Unikalny identyfikator zapasu
- Item Type: Typ zapasu:
    - FG (Finished Goods) - wyroby gotowe.
    - SFG (Semi-Finished Goods) - półprodukty.
    - RM (Raw Material) - surowce
- Group: Grupa produktowa (np. Structures, Beams, Plates, Modules).
- Description: Opis zapasu (np. "Steel Beam 12m").
- Unit of Measure: Jednostka miary (np. PCS).
- Model: model do jakiego należy dany zapas.
Relacje:
- Z sales_orders: Tylko zapasy z Item Type = FG mogą być używane w zamówieniach sprzedaży.
- Z routes: Dla każdego zapasu istnieje marszruta określająca proces produkcji.

### 2. Zbiór: workcenters (Gniazda robocze)
Opis: Zawiera informacje o gniazdach roboczych w pięciu wydziałach produkcyjnych.
Kolumny:
- Workcenter ID: Unikalny identyfikator gniazda roboczego (np. "WC-001").
- Department: Wydział, do którego należy gniazdo (np. Assembly, Welding, Machining, Painting, Packing).
- Workcenter Name: Unikalna nazwa gniazda (np. "Welding Station Alpha").
- Capacity: Liczba stanowisk w gnieździe roboczym (1-10).
- Employees per Station: Liczba pracowników przypadających na jedno stanowisko (najczęściej 1, sporadycznie 2-3).
- Efficiency: Efektywność gniazda roboczego (75-90%).
- Shifts: Liczba zmian (zawsze 2).
- Daily Availability (hours): Dzienne godziny dostępności (zawsze 16 godzin).
Relacje:
- Z routes: Operacje w marszrucie są przypisane do konkretnych gniazd roboczych na podstawie wydziału.

### 3. Zbiór: workcenter_availability (Dostępność gniazd roboczych)
Opis: Określa dzienną dostępność każdego gniazda roboczego przez dwa lata (2025-2026), z wyłączeniem weekendów i świąt.
Kolumny:
- Workcenter ID: Identyfikator gniazda (np. "WC-001").
- Date: Data dostępności (np. "2025-01-02").
- Start Time: Godzina rozpoczęcia pracy (zawsze "06:00").
- End Time: Godzina zakończenia pracy (zawsze "22:00").
- Daily Availability (hours): Łączna dzienna dostępność w godzinach (zawsze 16).
Relacje:
- Z routes: Informacje o dostępności mogą być używane do sprawdzania, czy gniazdo robocze jest dostępne w czasie przypisanej operacji.

### 4. Zbiór: routes (Marszruty)
Opis: Definiuje procesy produkcyjne dla wyrobów gotowych i półproduktów.
Kolumny:
- Item ID: Identyfikator zapasu (np. "FG-001", "SFG-005").
- Operation ID: Identyfikator operacji w marszrucie (np. "OP-001").
- Next Operation ID: Identyfikator następnej operacji (lub NULL dla ostatniej operacji).
- Previous Operation ID: Identyfikator poprzedniej operacji (lub NULL dla pierwszej operacji).
- Operation Name: Nazwa operacji zawierająca wydział i numer operacji w danym wydziale (np. "Welding 1").
- Workcenter ID: Gniazdo robocze przypisane do operacji.
- Setup Time (minutes): Czas przygotowawczy operacji.
- Run Time (minutes): Czas pracy operacji.
Relacje:
- Z inventory: Marszruty są przypisane do zapasów FG i SFG.
- Z workcenters: Operacje są przypisane do gniazd roboczych na podstawie wydziału.

### 5. Zbiór: customers (Klienci)
Opis: Zawiera informacje o klientach, do których kierowane są zamówienia sprzedaży.
Kolumny:
- Customer ID: Unikalny identyfikator klienta (np. "CUST-001").
- Customer Name: Nazwa klienta (np. "SteelCo AB").
- Country Code: Kod kraju klienta (PL, DE, SE).
- Address: Fikcyjny adres klienta, zawierający ulicę, miasto, kod pocztowy i kraj.
Relacje:
- Z sales_orders: Każde zamówienie sprzedaży jest przypisane do jednego z klientów.

### 6. Zbiór: sales_orders (Zamówienia sprzedaży)
Opis: Zawiera informacje o zamówieniach sprzedaży składanych przez klientów.
Kolumny:
- Order ID: Unikalny identyfikator zamówienia (np. "SO-0001").
- Customer ID: Identyfikator klienta przypisanego do zamówienia.
- Item ID: Identyfikator zapasu (FG), którego dotyczy zamówienie.
- Quantity: Ilość zamówionego produktu (1, 2 lub 4 sztuki).
- Delivery Date: Termin realizacji zamówienia (np. "2025-05-15").
Relacje:
- Z inventory: Zamówienia są przypisane tylko do zapasów FG.
- Z customers: Każde zamówienie przypisane jest do istniejącego klienta.

### 7. Zbiór: bom (Bill of Materials)
Opis: Zawiera informacje o materiałach lub produktach użytych do wytworzenia danego zapasu.
Kolumny:
- BOM ID: wyrób gotowy lub półprodukt do którego przypisane są komponenty.
- Component ID: Składnik używany w produkcji Item ID. Może być:
    - Wyrobem gotowym (FG) w kontekście struktury dla większego produktu.
    - Półproduktem (SFG).
    - Surowcem (RM).
- Quantity: Ilość komponentu wymagana do wyprodukowania jednej jednostki produktu 
- Unit of Measure: Jednostka miary komponentu (np. PCS).


### Relacje między zbiorami:
- inventory - routes: Każdy zapas ma przypisaną marszrutę produkcyjną.
- routes - workcenters: Operacje w marszrucie są przypisane do konkretnych gniazd roboczych.
- workcenters - workcenter_availability: Dostępność każdego gniazda jest określona w czasie.
- customers - sales_orders: Zamówienia są przypisane do klientów.
- inventory - sales_orders: Zamówienia dotyczą tylko wyrobów gotowych (FG).
